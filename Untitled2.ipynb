```python
import numpy as np

List1=[]
List2=[]
List3=[]
List4=[]

List = [List1, List2, List3, List4]

for i in range(4):
    for j in range(4):
        List[j].append(np.random.randint(10, 110))

Array=np.array(List)

print(Array,"\n")
        
print(Array[2,],"\n")

print(Array[:,1],"\n")

for i in range(4):
    print("Row ",i+1," sum: ",Array[i,:].sum())

for i in range(4):
    print("Row ",i+1," Max Value: ",Array[:,i].max())

List1 = [x if x < 50 else np.random.randint(10, 50) for x in List1]
List2 = [x if x < 50 else np.random.randint(10, 50) for x in List2]
List3 = [x if x < 50 else np.random.randint(10, 50) for x in List3]
List4 = [x if x < 50 else np.random.randint(10, 50) for x in List4]

List = [List1, List2, List3, List4]
Array=np.array(List)

print(Array,"\n")
```
